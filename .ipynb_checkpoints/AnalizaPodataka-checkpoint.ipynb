```python
import pandas as pd

# create a sample DataFrame
df = pd.DataFrame({'category': ['A', 'B', 'C', 'A', 'C', 'B']})

# apply one-hot encoding to the 'category' column and append the resulting columns to the original DataFrame
encoded_df = pd.get_dummies(df['category'], prefix='category')
df = pd.concat([df, encoded_df], axis=1)

# output the resulting DataFrame
print(df)
```
